```python
 # [1] 라이브러리 불러오기
import pandas as pd
import matplotlib.pyplot as plt

# [2] 데이터 불러오기 (Colab에 직접 업로드)
from google.colab import files
uploaded = files.upload()

df = pd.read_csv("owid-covid-data.csv")

# [3] 데이터 구조 확인
print(df.columns)
print(df[['location', 'date', 'total_cases', 'total_deaths']].head())

# [4] 특정 국가 예시: 대한민국
kr = df[df['location'] == 'South Korea']

# [5] 누적 확진자 및 사망자 시각화
plt.figure(figsize=(12, 6))
plt.plot(kr['date'], kr['total_cases'], label='Total Cases', color='blue')
plt.plot(kr['date'], kr['total_deaths'], label='Total Deaths', color='red')
plt.xticks(rotation=45)
plt.title("COVID-19 in South Korea")
plt.xlabel("Date")
plt.ylabel("Count")
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()
```
