```python
# 1. 기본 라이브러리
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 2. 데이터 업로드 (Colab)
from google.colab import files
uploaded = files.upload()

# 3. 데이터 불러오기
df = pd.read_csv("survey_results_public.csv")

# 4. 결측값 제거 및 기본 전처리
df_clean = df[['MainBranch', 'Age', 'EdLevel', 'RemoteWork', 'ConvertedCompYearly']].dropna()

# 5. 연봉 분포 시각화 (이상치 제외)
df_clean = df_clean[df_clean['ConvertedCompYearly'] < 500000]

plt.figure(figsize=(10, 6))
sns.histplot(df_clean['ConvertedCompYearly'], bins=50, kde=True)
plt.title("연봉 분포 (USD, 상위 50만 이하)")
plt.xlabel("연봉 ($)")
plt.ylabel("응답자 수")
plt.grid(True)
plt.show()

# 6. 원격근무와 연봉 관계
plt.figure(figsize=(10, 6))
sns.boxplot(x='RemoteWork', y='ConvertedCompYearly', data=df_clean)
plt.xticks(rotation=20)
plt.title("원격근무 여부와 연봉 관계")
plt.xlabel("Remote Work")
plt.ylabel("연봉 ($)")
plt.grid(True)
plt.show()

# 7. 학력별 평균 연봉
edu_salary = df_clean.groupby('EdLevel')['ConvertedCompYearly'].mean().sort_values(ascending=False)
edu_salary.plot(kind='barh', figsize=(10,6), color='skyblue')
plt.title("학력별 평균 연봉")
plt.xlabel("평균 연봉 ($)")
plt.ylabel("학력 수준")
plt.grid(axis='x')
plt.tight_layout()
plt.show()
```
