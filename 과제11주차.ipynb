```python
# 1. 라이브러리 로딩
import pandas as pd
import matplotlib.pyplot as plt

# 2. 데이터 불러오기
from google.colab import files
uploaded = files.upload()

# Load only the uploaded file
hi = pd.read_csv("hawaii-coviddata.csv")

# 3. 날짜 컬럼 datetime 변환
hi['end_date'] = pd.to_datetime(hi['end_date'])

# 4. 주별 확진자 추세 그래프 (Hawaii only)
plt.figure(figsize=(12, 6))
plt.plot(hi['end_date'], hi['new_cases'], label='Hawaii')
plt.xlabel('Date')
plt.ylabel('New Cases')
plt.title('Weekly New COVID-19 Cases in Hawaii')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# 5. 주별 사망자 추세 그래프 (Hawaii only)
plt.figure(figsize=(12, 6))
plt.plot(hi['end_date'], hi['new_deaths'], label='Hawaii')
plt.xlabel('Date')
plt.ylabel('New Deaths')
plt.title('Weekly New COVID-19 Deaths in Hawaii')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
