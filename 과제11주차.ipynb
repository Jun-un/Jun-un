```python
# 1.  
import pandas as pd
import matplotlib.pyplot as plt

# 2.  
from google.colab import files
uploaded = files.upload()

# Load only the uploaded file
hi = pd.read_csv("hawaii-coviddata.csv")

# 3.   datetime 
hi['end_date'] = pd.to_datetime(hi['end_date'])

# 4.     (Hawaii only)
plt.figure(figsize=(12, 6))
plt.plot(hi['end_date'], hi['new_cases'], label='Hawaii')
plt.xlabel('Date')
plt.ylabel('New Cases')
plt.title('Weekly New COVID-19 Cases in Hawaii')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# 5.     (Hawaii only)
plt.figure(figsize=(12, 6))
plt.plot(hi['end_date'], hi['new_deaths'], label='Hawaii')
plt.xlabel('Date')
plt.ylabel('New Deaths')
plt.title('Weekly New COVID-19 Deaths in Hawaii')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
